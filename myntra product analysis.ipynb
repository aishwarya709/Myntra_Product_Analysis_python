```python
#import Required Libraries

import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Create Myntra Dataset

data = {
    "Product_Name": [
        "Men T-Shirt", "Women Dress", "Jeans", "Kurti",
        "Men Shirt", "Handbag", "Jacket", "Top"
    ],
    "Brand": [
        "Puma", "Zara", "Levis", "Biba",
        "H&M", "Lavie", "Roadster", "Forever21"
    ],
    "Category": [
        "Clothing", "Clothing", "Clothing", "Ethnic Wear",
        "Clothing", "Accessories", "Clothing", "Clothing"
    ],
    "Price": [799, 1999, 2499, 1599, 1299, 2999, 2799, 999],
    "Rating": [4.2, 4.5, 4.3, 4.4, 4.1, 4.0, 4.2, 4.3],
    "Reviews": [1200, 980, 1500, 870, 640, 430, 900, 780]
}

df = pd.DataFrame(data)
df
```

```python
#Basic Data Analysis

df.info()
df.describe()
```

```python
#Category-wise Product Count

category_count = df["Category"].value_counts()
category_count
```

```python
#Visualization

category_count.plot(kind="bar")
plt.title("Number of Products per Category (Myntra)")
plt.xlabel("Category")
plt.ylabel("Count")
plt.show()
```

```python
#Average Rating per Category

avg_rating = df.groupby("Category")["Rating"].mean()
avg_rating
```

```python
#Average Price per Category

avg_price = df.groupby("Category")["Price"].mean()
avg_price
```

```python
#Top Rated Products

top_rated = df.sort_values(by="Rating", ascending=False).head(5)
top_rated[["Product_Name", "Rating", "Price"]]
```

```python
#Most Reviewed Products

top_reviewed = df.sort_values(by="Reviews", ascending=False).head(5)
top_reviewed[["Product_Name", "Reviews"]]
```

```python
#Key Business Insights

print("Most common category:", category_count.idxmax())
print("Highest rated category:", avg_rating.idxmax())
print("Most expensive category:", avg_price.idxmax())
```
